# 結果の可視化

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn  as sns
```

```python
def read_result(path):
    ret = []
    with open(path, "r") as r_f:
        for line in r_f:
            ret.append(float(line.rstrip()))
    return np.array(ret)
```

```python
def read_q(path):
    ret = []
    with open(path, "r") as r_f:
        for line in r_f:
            fc, fd = line.rstrip().split(",")
            
            fc = float(fc)
            fd = float(fd)
            
            ret.append(fc)
            ret.append(fd)
            
    return np.array(ret)
```

## 協調確率の可視化

```python
f1 = "./graph/q/random/prisoners_dilemma_coop.csv"
f2 = "./observable/q/random/prisoners_dilemma_coop.csv"

d = {"none":f1, "observable":f2}

d["iteration"] = np.arange(1000)

df = pd.DataFrame(data = d)
```

```python
current_palette = sns.color_palette("hls", 2)
plt.figure(figsize=(10, 5))

for i, label in enumerate(d):
    plt.subplot(1, 2, 1+i)
    sns.regplot(x="iteration", y=label, data=df, order=2, scatter_kws={'alpha':0.4}, marker=".", label=label, color=current_palette[i])
    plt.ylabel("cooperation ratio")
    plt.legend(fontsize=14)
    
plt.show()
```

### Q学習

```python
prefix = "./observable/q/"
file_name = "/prisoners_dilemma_coop.csv"
files = ["powerlaw_cluster", "complete", "grid2d", "inner_dence_clustered", "multiple_clustered", "one_dim_regular", "random"]
d = dict([(f, read_result(prefix+f+file_name)) for f in files])

d["iteration"] = np.arange(1000)

df = pd.DataFrame(data = d)
```

```python
current_palette = sns.color_palette("hls", (len(files)))
plt.figure(figsize=(20, 8))

for i, label in enumerate(files):
    plt.subplot(241+i)
    sns.regplot(x="iteration", y=label, data=df, order=2, scatter_kws={'alpha':0.4}, marker=".", label=label, color=current_palette[i])
    plt.ylabel("cooperation ratio")
    plt.ylim((0,1.0))
    plt.legend(fontsize=14)
    
plt.show()
```
