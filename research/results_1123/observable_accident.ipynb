# 結果の可視化

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn  as sns
```

```python
def read_result(path):
    ret = []
    with open(path, "r") as r_f:
        for line in r_f:
            ret.append(float(line.rstrip()))
    return np.array(ret)
```

```python
def read_q(path):
    ret = []
    with open(path, "r") as r_f:
        for line in r_f:
            fc, fd = line.rstrip().split(",")
            
            fc = float(fc)
            fd = float(fd)
            
            ret.append(fc)
            ret.append(fd)
            
    return np.array(ret)
```

## 協調確率の可視化

### Q学習

```python
file_name = "./observable_accident/random/q/prisoners_kaishou_0_coop35.csv"
d = {"observable": read_result(file_name)}

d["iteration"] = np.arange(1000)

df = pd.DataFrame(data = d)
```

```python
current_palette = sns.color_palette("hls", 1)
plt.figure(figsize=(7, 5))


sns.regplot(x="iteration", y="observable", data=df, order=2, scatter_kws={'alpha':0.4}, line_kws={'alpha':0.0}, marker=".", label="observable", color=current_palette[0])
plt.ylabel("cooperation ratio")
plt.ylim((0,1.0))
plt.legend(fontsize=14)
    
plt.show()
```
