# 結果の可視化

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn  as sns
```

```python
def read_result(path):
    ret = []
    with open(path, "r") as r_f:
        for line in r_f:
            ret.append(float(line.rstrip()))
    return np.array(ret)
```

```python
def read_q(path):
    ret = []
    with open(path, "r") as r_f:
        for line in r_f:
            fc, fd = line.rstrip().split(",")
            
            fc = float(fc)
            fd = float(fd)
            
            ret.append(fc)
            ret.append(fd)
            
    return np.array(ret)
```

## 協調確率の可視化

```python
prefix = "./share_q/q/"
file_name = "prisoners_dilemma_50.0"
files = ["powerlaw_cluster", "complete", "grid2d", "inner_dence_clustered", "multiple_clustered", "one_dim_regular", "random"]
d = dict([(f, read_result(prefix+f+"/"+file_name+"_coop.csv")) for f in files])

d["iteration"] = np.arange(1000)

df = pd.DataFrame(data = d)
```

```python
current_palette = sns.color_palette("hls", (len(files)))
plt.figure(figsize=(20, 15))

for i, label in enumerate(files):
    plt.subplot(3,4,i+1)
    sns.regplot(x="iteration", y=label, data=df, order=2, scatter_kws={'alpha':0.4}, marker=".", label=label, color=current_palette[i])
    plt.ylim((0,0.6))
    plt.ylabel("cooperation ratio")
    plt.legend(fontsize=14)
    
plt.show()
```

## Q値の可視化

```python
prefix = "./share_q/q/"
file_name = "prisoners_dilemma_50.0"
files = ["powerlaw_cluster", "complete", "grid2d", "inner_dence_clustered", "multiple_clustered", "one_dim_regular", "random"]
d = dict([(f, read_q(prefix+f+"/"+file_name+"_q.csv")) for f in files])

iteration = []
for i in range(1000):
    iteration.append(i)
    iteration.append(i)
    
d["iteration"] = np.array(iteration)
d["type"] = np.array(["c", "d"] * 1000)

df = pd.DataFrame(data = d)
df_c = df[df["type"] == "c"]
df_d = df[df["type"] == "d"]

df_c.index = df_c["iteration"]
df_d.index = df_d["iteration"]
```

```python
plt.figure(figsize=(20, 15))

for i, label in enumerate(files):
    plt.subplot(3, 4, 1+i)
    plt.scatter(df_c["iteration"], df_c[label], color="lightcoral", marker=".", alpha=0.2)
    plt.scatter(df_d["iteration"], df_d[label], color="dodgerblue", marker=".", alpha=0.2)
    plt.ylabel("q value")
    plt.title(label)
    
plt.show()
```
