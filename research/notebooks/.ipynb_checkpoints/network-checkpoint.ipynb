# ネットワークの特性調査

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

### 比較するネットワーク

```python
n_node = 100
r_graph = nx.random_regular_graph(30, n_node)
ba30_graph = nx.barabasi_albert_graph(n_node, 30)
ba50_graph = nx.barabasi_albert_graph(n_node, 50)
ba70_graph = nx.barabasi_albert_graph(n_node, 70)
ws3003_graph = nx.watts_strogatz_graph(n_node, 30, 0.3)
ws3007_graph = nx.watts_strogatz_graph(n_node, 30, 0.7)
ws7003_graph = nx.watts_strogatz_graph(n_node, 70, 0.3)
ws7007_graph = nx.watts_strogatz_graph(n_node, 70, 0.7)
complete_graph = nx.complete_graph(n_node)
```

###  基礎情報

#### 平均経路長
ネットワーク上における任意のノード間の最短経路長を平均した値
#### クラスタ係数
クラスタ係数はネットワークの局所的な凝縮性を示す指標であり，「あるノードの隣接ノード間にリンクが存在する確率」として定義される
#### 次数相関
各リンクの終点であるノード間の次数相関を示す指標で，-1 <= r <=1の範囲の値をとる．  
次数相関rが正の値をとるとき，次数が高いノード同士が接続されている傾向があることを示し,  
rが負の値をとるときは次数が高いノードと低いノードが接続されている傾向があることを示す．

```python
def summary(G):
    """
    ネットワークの基本情報
    """
    print(nx.info(G))
    print("平均経路長", nx.average_shortest_path_length(G))
    print("クラスタ係数", nx.average_clustering(G))
    print("次数相関", nx.degree_assortativity_coefficient(G))
    degree_distribution(G)
```

```python
def degree_distribution(G):
    """
    ネットワークと次数プロット
    """
    degree_sequence = sorted(nx.degree(G).values(),reverse=True) # degree sequence
    #print "Degree sequence", degree_sequence
    dmax = max(degree_sequence)

    plt.loglog(degree_sequence,'b-',marker='o')
    plt.title("Degree rank plot")
    plt.ylabel("degree")
    plt.xlabel("rank")

    # draw graph in inset
    d = nx.degree(G)
    plt.axes([0.45,0.45,0.45,0.45])
    Gcc = sorted(nx.connected_component_subgraphs(G), key = len, reverse=True)[0]
    pos = nx.circular_layout(Gcc)
    plt.axis('off')
    nx.draw_networkx_nodes(Gcc, pos, alpha=0.5, node_size=[v*0.8 for v in d.values()])
    nx.draw_networkx_edges(Gcc, pos, alpha=0.4, width=0.1)

    plt.show()
```

```python
summary(r_graph)
```
